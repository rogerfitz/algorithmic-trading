```python
import pandas as pd
import pandas_datareader.data as web
import datetime
import statsmodels.formula.api as smf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start=datetime.datetime(2006,5,24)
end=datetime.datetime(2012,4,9)
```

```python
USO=web.DataReader("USO",'yahoo', start,end)
GLD=web.DataReader("GLD",'yahoo', start,end)
```

```python
(USO/GLD)['Close'].plot()
```

```python
LOOKBACK=20
headRatio=[]
```

```python
X=USO['Close'].tolist()
y=GLD['Close'].tolist()
length=len(X)
dat=pd.DataFrame({"USO": X, "GLD": y})
hedgeRatio=np.full(length, np.NaN)
spread=np.full(length, np.NaN)
```

```python
for t in range(LOOKBACK,length):
    window=dat.iloc[t-LOOKBACK:t]
    results=smf.ols('USO~GLD',data=window).fit()
    b0,b1=results.params
    hedgeRatio[t]=b1
    spread[t]=(dat.iloc[t]['USO']-b1*dat.iloc[t]['GLD'])
```

```python
plt.plot(spread)
```
